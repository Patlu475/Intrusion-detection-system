## 7. Display Figures

```python
from IPython.display import Image, display
import os

for dataset in ['nsl-kdd', 'unsw-nb15']:
    print(f'\n{"="*60}')
    print(f'{dataset.upper()} FIGURES')
    print(f'{"="*60}')
    fig_dir = f'figures/{dataset}'
    if os.path.exists(fig_dir):
        for fname in sorted(os.listdir(fig_dir)):
            if fname.endswith('.png'):
                print(f'\n--- {fname} ---')
                display(Image(filename=f'{fig_dir}/{fname}'))
```
